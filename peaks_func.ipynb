# Plot the Peaks Function

[Reference in MATLAB](https://www.mathworks.com/help/matlab/ref/peaks.html#mw_46aeee28-390e-4373-aa47-e4a52447fc85)

https://plotly.com/matlab/

$ z = 3(1-x)^2 e^{-x^2 - (y+1)^2} - 10( \frac{x}{5} - x^3 -y^5) e^{-x^2 - y^2} - \frac{1}{3} e^{-(x+1)^2 - y^2} $


```python
import numpy as np
```

```python
def peaks(x, y):
    p1 = 3 * (1-x)**2 * np.exp( -x**2 - (y+1)**2 ) 
    p2 = 10 * ( x/5 - np.power(x, 3) - np.power(y, 5) ) * np.exp( -x**2 - y**2 ) 
    p3 = (1/3) * np.exp( -(x+1)**2 - y**2 ) 
    z = (p1 - p2 - p3) / 5
    return z 
```

```python
range = [-5, 5]
spacing = 0.1 
```

```python
x = np.linspace(range[0], range[1], int((range[1] - range[0])/spacing)+1 )
print(x.shape, '\n', x)
```

```python
y = np.linspace(range[0], range[1], int((range[1] - range[0])/spacing)+1 )
print(y.shape, '\n', y)
```

```python
xv, yv = np.meshgrid(x, y)
xv
```

```python
xv.shape
```

```python
yv
```

```python
yv.shape
```

```python
z = peaks(xv, yv)
z 
```

```python
z.shape
```

```python
import plotly.graph_objects as go 

fig = go.Figure(
    data=[go.Surface(
        z=z, 
        x=x,
        y=y,
    )]
) 

fig.update_traces(
    contours_z = dict(
        show=True, 
        usecolormap=True, 
        highlightcolor="limegreen", 
        project_z=True, 
    )
)

fig.update_layout(
    autosize=False, 
    width=800, 
    height=800,
    scene= dict(
        xaxis = dict(nticks=11, range=range,),
        yaxis = dict(nticks=11, range=range,),
        zaxis = dict(nticks=11, range=range,), 
    ), 
    scene_aspectmode='cube', 
)

fig.show()
```
